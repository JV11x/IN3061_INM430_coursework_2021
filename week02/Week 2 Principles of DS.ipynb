```python
# EXERCISE 3
# STEP 1
# Check dataset summary
WHO = pd.read_csv("TB_burden_countries_2014-09-29(1).csv")
WHO.describe()
```

```python
# Check missing varriables
WHO.isna().sum()
```

```python
# STEP 2
#removing all columns where more than %10 of the data is missing 
WHO_null = WHO.isnull().sum() / len(WHO)
missing_features = WHO_null[WHO_null > 0.1].index
WHO.drop(missing_features, axis=1, inplace=True)
```

```python
# using fillna to replace missing columns with mean values
WHOre = WHO.fillna(WHO.mean())
```

```python
WHOre.isna().sum()
```

```python
# STEP 3
# Plotting Histograms of various earning metrics
plt.subplot(221)
plt.hist(WHOre['e_prev_100k_hi'])
plt.title('e_prev_100k_hi')
plt.subplot(222)
plt.title('e_prev_100k_lo')
plt.hist(WHOre['e_prev_100k_lo'])
plt.subplot(223)
plt.title('e_prev_100k')
plt.hist(WHOre['e_prev_100k'])
plt.show
```

```python
# STEP 4
# Using a log function to reduce te histogram skew
E100kWHOre = np.log(WHOre['e_prev_100k'])
E100kloWHOre = np.log(WHOre['e_prev_100k_lo'])
E100khiWHOre = np.log(WHOre['e_prev_100k_hi'])
```

```python
plt.subplot(221)
plt.hist(E100khiWHOre)
plt.title('e_prev_100k_hi')
plt.subplot(222)
plt.title('e_prev_100k_lo')
plt.hist(E100kloWHOre)
plt.subplot(223)
plt.title('e_prev_100k')
plt.hist(E100kWHOre)
plt.show
```

```python
# Creating DF with only numeric values
WHON = WHOre.select_dtypes(include='number')
```

```python
# STEP 5
# Normalising data so all values fall between 0 and 1 to make means easier to interept 
def NormalizeData(data):
    return (data - np.min(data)) / (np.max(data) - np.min(data))

SWHON = NormalizeData(WHON)

SWHON.describe()
```

```python
# STEP 5
# Visualising means of columns
WM = SWHON.mean(axis=0)
WSD = SWHON.std(axis=0)

WM.plot(x="X", y=[], kind="bar")
```
