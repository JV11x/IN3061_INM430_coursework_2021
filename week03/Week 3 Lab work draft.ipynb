```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import scipy.stats as stats

cars = pd.read_csv('accord_sedan.csv')

cars.describe()
```

```python
cars.isna().sum()
```

```python
f, (ax1, ax2) = plt.subplots(1, 2)
ax1.boxplot(cars['price'])
ax1.set_title('Price')
ax2.boxplot(cars['mileage'])
ax2.set_title('Mileage')
```

```python
f, (ax1, ax2) = plt.subplots(1, 2)
ax1.hist(cars['price'])
ax1.set_title('Price')
ax2.hist(cars['mileage'])
ax2.set_title('Mileage')
```

```python
plt.scatter(cars['price'], cars['mileage'])
plt.show
```

```python
pricecars = cars['price']
mileagecars = cars['mileage']
```

```python
pricemean = np.mean(cars['price'])
stdP = np.std(cars['price'])
mileageMean = np.mean(cars['mileage'])
stdM = np.std(cars['mileage'])
```

```python
for i, val in enumerate (cars['mileage']):
    if abs(val - mileageMean) > 2.2*stdM:
        cars.loc[i, 'isOutlierMileage'] = 1
```

```python
for i, val in enumerate (cars['price']):
    if abs(val - pricemean) > 2.2*stdP:
        cars.loc[i, 'isOutlierPrice'] = 1
        
```

```python
plt.scatter(cars['price'],cars['mileage'], c= cars.isOutlierMileage)
plt.scatter(cars['price'],cars['mileage'], c= cars.isOutlierPrice)
plt.scatter(cars['price'],cars['mileage'], alpha= 0.5)
plt.show
```
