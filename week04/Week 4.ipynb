```python
heart = pd.read_csv('heart.csv')
heart
```

```python
heart['sex'].replace(0, 'Female',inplace=True)
heart['sex'].replace(1, 'Male',inplace=True)

heart.rename(columns={'sex':'gender'}, inplace = True)

```

```python
heart['target'].replace(0, 'Disease',inplace=True)
heart['target'].replace(1, 'Clear',inplace=True)
heart.rename(columns={'target':'hasHeartDisease'}, inplace = True)

heart
```

```python
heart.describe()
```

```python
print(heart.mean())
print(heart.std())
```

```python
haveHeart = heart.loc[(heart['hasHeartDisease'] == 'Disease')]
clearHeart = heart.loc[(heart['hasHeartDisease'] == 'Clear')]
haveHeart.describe()
```

```python
print('mean heart rate of disease', haveHeart.trestbps.mean())
print('mean heart rate of clear',clearHeart.trestbps.mean())
print('stdv heart rate of disease', haveHeart.trestbps.std())
print('stdv heart rate of clear',clearHeart.trestbps.std())
```

```python
f, (ax1, ax2) = plt.subplots(ncols=2,sharey=True)
ax1.boxplot(haveHeart['trestbps'])
ax1.set_title('Have Heart BPS')
ax2.boxplot(clearHeart['trestbps'])
ax2.set_title('Clear Heart BPS')
```

```python
f, (ax1, ax2) = plt.subplots(ncols=2,sharey=True)
ax1.hist(haveHeart['trestbps'])
ax1.set_title('Have Heart BPS')
ax2.hist(clearHeart['trestbps'])
ax2.set_title('Clear Heart BPS')
```

```python
stats.ttest_ind(haveHeart['trestbps'],clearHeart['trestbps'])
```

```python
# function to calculate Cohen's d for independent samples
def cohend(d1, d2):
# calculate the size of samples
    n1, n2 = len(d1), len(d2)
# calculate the variance of the samples
    s1, s2 = var(d1, ddof=1), var(d2, ddof=1)
# calculate the pooled standard deviation
    s = sqrt(((n1 - 1) * s1 + (n2 - 1) * s2) / (n1 + n2 - 2))
# calculate the means of the samples
    u1, u2 = mean(d1), mean(d2)
# calculate the effect size
    return (u1 - u2) / s
```

```python
cohend(haveHeart['trestbps'],clearHeart['trestbps'])
```

The effect size is relatively small
